```python
import pandas as pd
data = pd.read_csv('slr.csv')
data.head()
```

```python
data.describe()
```

```python
x= data[['chirps per second']]
y = data['temperature (F)']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=34,test_size=0.25)
```

```python
data.shape
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.linear_model import LinearRegression
slr_model = LinearRegression()  
```

```python
slr_model.fit(x_train,y_train)
```

```python
slope = slr_model.coef_
intercept = slr_model.intercept_
print('slope:',slope)
print('intercept:',intercept)
```

```python
y_pred_train = slr_model.predict(x_train)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
r2_train = r2_score(y_train,y_pred_train)
mse_train = mean_squared_error(y_train,y_pred_train)
print('r2_train:',r2_train)
print('mse_train:',mse_train)
```

```python
y_pred_test = slr_model.predict(x_test)
```

```python
r2_test = r2_score(y_test,y_pred_test)
mse_test = mean_squared_error(y_test,y_pred_test)
print('r2_test:',r2_test)
print('mse_test:',mse_test)
```

```python
compare = pd.DataFrame({'Actual':y_test,'Predicted':y_pred_test})
compare.head()
```

```python
compare.to_csv('compare.csv',index=True)
```

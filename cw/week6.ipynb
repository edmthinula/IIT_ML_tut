```python
# Define function to find outliers using IQR method
def find_outliers_IQR(data):
    # Calculate first quartile
    q1 = data.quantile(0.25)
    # Calculate third quartile
    q3 = data.quantile(0.75)
    # Calculate interquartile range
    IQR = q3-q1
    # Find outliers using 1.5 * IQR rule
    outliers = data[((data<(q1-1.5*IQR))|(data>(q3+1.5*IQR)))]
    return outliers
```

Code Reuse Session 1

```python
# Remove age outliers from the dataset
age = find_outliers_IQR(data['Age']).index
data.drop(index=age, inplace=True)
survival_months = find_outliers_IQR(data['Survival_Months']).index
data.drop(index=survival_months, inplace=True)
```

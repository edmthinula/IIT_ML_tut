```python
def find_outliers_IQR(data):

  q1=data.quantile(0.25)

  q3=data.quantile(0.75)

  IQR=q3-q1

  outliers = data[((data<(q1-1.5*IQR))|(data> (q3+1.5*IQR)))]

  return outliers
```

```python

# List of target columns
outlier_columns = ["Regional_Node_Examined", "Age", "Tumor_Size", "Reginol_Node_Positive", "Survival_Months"]

# Identify outliers for each column
outlier_mask = data[outlier_columns].apply(find_outliers_IQR)

# Count number of outliers per row
outlier_count_per_row = outlier_mask.sum(axis=1)
data.head(10)
```

```python
def find_outliers_IQR(data):
  q1=data.quantile(0.25)
  q3=data.quantile(0.75)
  IQR=q3-q1
  outliers = data[((data<(q1-1.5*IQR)) | (data>(q3+1.5*IQR)))]
  return outliers
```

```python
for i in data.columns:
    age_outliers = find_outliers_IQR(data[i])  # Pass the actual column data
    print("Column: " + str(i))
    print("number of outliers: " + str(age_outliers.count()))  # Use count() for non-NaN values
    print("max outlier value: " + str(age_outliers.max()))
    print("min outlier value: " + str(age_outliers.min()))
    print(age_outliers.sort_values(ascending=False))
```

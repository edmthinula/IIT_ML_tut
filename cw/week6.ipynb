```python
data.drop(['Occupation'], axis=1,inplace=True)
```

```python
data = data.dropna()
```

```python
data = data.astype({"Tumor_Size": "int", "Regional_Node_Examined": "int","Age": "int"})
```

```python
data.head(10)
```

```python
def find_outliers_IQR(data):

  q1=data.quantile(0.25)

  q3=data.quantile(0.75)

  IQR=q3-q1

  outliers = data[((data<(q1-1.5*IQR))|(data> (q3+1.5*IQR)))]

  return outliers
```

```python

# List of target columns
outlier_columns = ["Regional_Node_Examined", "Age", "Tumor_Size", "Reginol_Node_Positive", "Survival_Months"]

# Identify outliers for each column
outlier_mask = data[outlier_columns].apply(find_outliers_IQR)

# Count number of outliers per row
outlier_count_per_row = outlier_mask.sum(axis=1)
data.head(10)
```

```python
data[outlier_count_per_row.reindex(data.index, fill_value=0) < 2]
data
```

```python
data['T_Stage'] = labale_encoder.fit_transform(data['T_Stage'])
data['N_Stage'] = labale_encoder.fit_transform(data['N_Stage'])
data['6th_Stage'] = labale_encoder.fit_transform(data['6th_Stage'])
data['Differentiated']= labale_encoder.fit_transform(data['Differentiated'])
data['A_Stage']= labale_encoder.fit_transform(data['A_Stage'])
data['Estrogen_Status']= labale_encoder.fit_transform(data['Estrogen_Status'])
data['Progesterone_Status']= labale_encoder.fit_transform(data['Progesterone_Status'])
data['Mortality_Status'] = labale_encoder.fit_transform(data['Mortality_Status'])
```
